```python
import numpy as np
import matplotlib.pyplot as plt

#    
def simulate_two_envelopes(num_trials):
    gains_if_switch = []  #     
    
    for _ in range(num_trials):
        #     : X  2*X
        X = np.random.randint(1, 101)  #    
        envelope = np.random.choice([X, 2*X])  #   
        
        #  ,   
        if envelope == X:
            gain = 2*X
        else:
            gain = X
        gains_if_switch.append(gain)
    
    #      
    return np.mean(gains_if_switch)

#     
num_trials_values = np.arange(100, 10001, 100)  #  100  10000
expected_gains = [simulate_two_envelopes(n) for n in num_trials_values]

#  
plt.figure(figsize=(10,6))
plt.plot(num_trials_values, expected_gains, 'b-', linewidth=2)
plt.xlabel(" ")
plt.ylabel("   ")
plt.title("  ")
plt.grid(True, alpha=0.3)
plt.axhline(y=1.5 * 50, color='r', linestyle='--', alpha=0.7, label=' ')
plt.legend()
plt.show()
```
